```python
import pandas as pd

file1 = "purchase_data.csv"
df = pd.read_csv(file1)

df.head()
```

```python
player_count = df.SN.nunique()

print("Unique Players: " + str(player_count))
```

```python
unique_item_count = df['Item ID'].nunique()
avg_price = df.Price.mean()
total_item_count = df['Item ID'].count()
total_rev = df.Price.sum()

summary_df = pd.DataFrame({
    
    'Number of Unique Items': unique_item_count,
    'Average Price': avg_price,
    'Number of Purchases': total_item_count,
    'Total Revenue': total_rev
    
}, index = [0])

summary_df['Average Price'] = summary_df['Average Price'].map("${0:,.2f}".format)
summary_df['Total Revenue'] = summary_df['Total Revenue'].map("${0:,.2f}".format)

summary_df
```

```python
df1 = df.drop_duplicates('SN')

group_df = df1.groupby('Gender').count()

group_df = pd.DataFrame(group_df['Purchase ID'].rename('Total Players'))

group_df['Percentage of Players'] = group_df['Total Players'] / group_df['Total Players'].sum()
group_df['Percentage of Players'] = group_df['Percentage of Players'] * 100
group_df['Percentage of Players'] = group_df['Percentage of Players'].map("{:,.2f}%".format)

group_df
```

```python
group_df2 = df.groupby('Gender')

group_df2 = pd.DataFrame({
    'Purchase Count': group_df2['Purchase ID'].count(),
    'Average Purchase Price': group_df2['Price'].mean(),
    'Total Purchase Value': group_df2['Price'].sum(),
    'Average Total Purchase per Person': (group_df2['Price'].sum() / group_df['Total Players'])
})

group_df2['Average Purchase Price'] = group_df2['Average Purchase Price'].map("${0:,.2f}".format)
group_df2['Total Purchase Value'] = group_df2['Total Purchase Value'].map("${0:,.2f}".format)
group_df2['Average Total Purchase per Person'] = group_df2['Average Total Purchase per Person'].map("${0:,.2f}".format)

group_df2
```

```python
bins = [0,9,14,19,24,29,34,39,100]
bin_labels = ['<10','10-14','15-19','20-24','25-29','30-34','35-39','40+']

df1['Age Group'] = pd.cut(df1['Age'],bins=bins, labels = bin_labels)
df['Age Group'] = pd.cut(df['Age'],bins=bins, labels = bin_labels)

df_age_group = df1.groupby('Age Group').count()
df_age_group = pd.DataFrame(df_age_group['Purchase ID'].rename('Total Count'))

df_age_group['Percentage of Players'] = df_age_group['Total Count'] / df_age_group['Total Count'].sum()
df_age_group['Percentage of Players'] = df_age_group['Percentage of Players'] * 100
df_age_group['Percentage of Players'] = df_age_group['Percentage of Players'].map("{:,.2f}%".format)

df_age_group
```

```python
group_df3 = df.groupby('Age Group')

df_age_group2 = pd.DataFrame({
    'Purchase Count': group_df3['Age Group'].count(),
    'Average Purchase Price': group_df3['Price'].mean(),
    'Total Purchase Value': group_df3['Price'].sum(),
    'Average Total Purchase per Person': (group_df3['Price'].sum() / df_age_group['Total Count'])
})

df_age_group2['Average Purchase Price'] = df_age_group2['Average Purchase Price'].map("${0:,.2f}".format)
df_age_group2['Total Purchase Value'] = df_age_group2['Total Purchase Value'].map("${0:,.2f}".format)
df_age_group2['Average Total Purchase per Person'] = df_age_group2['Average Total Purchase per Person'].map("${0:,.2f}".format)

df_age_group2
```

```python
top_spender_df = df.groupby('SN')

top_spender_df2 = pd.DataFrame({
    'Purchase Count': top_spender_df['Price'].count(),
    'Average Purchase Price': top_spender_df['Price'].sum() / top_spender_df['Price'].count(),
    'Total Purchase Value': top_spender_df['Price'].sum()
})

top_spender_df2 = top_spender_df2.sort_values('Total Purchase Value',ascending = False)
top_spender_df2['Total Purchase Value'] = top_spender_df2['Total Purchase Value'].map("${0:,.2f}".format)
top_spender_df2['Average Purchase Price'] = top_spender_df2['Average Purchase Price'].map("${0:,.2f}".format)

top_spender_df2.head()
```

```python
pop_items_df = df[['Item ID','Item Name','Price']]

pop_items_df = pop_items_df.groupby(['Item ID','Item Name'])

pop_items_df = pd.DataFrame({
    'Purchase Count':pop_items_df['Item ID'].count(),
    'Average Purchase Price':pop_items_df['Price'].mean(),
    'Total Purchase Value':pop_items_df['Item ID'].count() * pop_items_df['Price'].mean()
})

pop_items_df = pop_items_df.sort_values('Purchase Count',ascending = False)

pop_items_df['Total Purchase Value'] = pop_items_df['Total Purchase Value'].map("${0:,.2f}".format)
pop_items_df['Average Purchase Price'] = pop_items_df['Average Purchase Price'].map("${0:,.2f}".format)

pop_items_df.head()
```

```python
pop_items_df2 = df[['Item ID','Item Name','Price']]

pop_items_df2 = pop_items_df2.groupby(['Item ID','Item Name'])

pop_items_df2 = pd.DataFrame({
    'Purchase Count':pop_items_df2['Item ID'].count(),
    'Average Purchase Price':pop_items_df2['Price'].mean(),
    'Total Purchase Value':pop_items_df2['Item ID'].count() * pop_items_df2['Price'].mean()
})

pop_items_df2 = pop_items_df2.sort_values('Total Purchase Value',ascending = False)

pop_items_df2['Total Purchase Value'] = pop_items_df2['Total Purchase Value'].map("${0:,.2f}".format)
pop_items_df2['Average Purchase Price'] = pop_items_df2['Average Purchase Price'].map("${0:,.2f}".format)

pop_items_df2.head()
```
